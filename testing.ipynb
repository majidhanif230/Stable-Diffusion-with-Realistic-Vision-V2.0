# Import Libraries

```python
# Step 1: Import necessary libraries
import os
import torch
from PIL import Image
from diffusers import StableDiffusionPipeline
from huggingface_hub import login
```

```python
login('add your key') # add you login key
```

# Load Model

```python
from diffusers import StableDiffusionPipeline

# Load the model
model_id = "majid230/Realistic_Vision_V2.0"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")
```

# Sample Work Case 1

```python
# Step 4: Create results directory
os.makedirs('results', exist_ok=True)

# Step 5: Define your prompt
prompt = "A 24 years handsome boy name Majid Hanif wearing trouser shirt at big mall,portrait,ultra realistic,realism, full face,4k. "

# Step 6: Generate the image
with torch.no_grad():
    generated_image = pipe(prompt).images[0]

# Step 7: Save the generated image
generated_image.save("results/Majid_Hanif2.png")
print("Images Generated")
```
